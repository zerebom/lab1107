```python
sns.set_palette("husl")
sns.set_style('white')
sns.set_context('notebook',font_scale=1.1)
cols=df.columns[:-1]
fig,axs=plt.subplots(5,4,figsize=(15,12))
for i,col in enumerate(cols):
    ax=axs[i%5,i//5]
    sns.distplot(df[col],ax=ax, norm_hist=False)
    ax.set_xlabel(col,fontsize=18)
plt.tight_layout()
plt.savefig('./hist_by_label1010.png')

plt.show()
```

```python
count_cols=[col for col in df.columns if 'count' in col]
count_df=df[count_cols]
```

```python
count_df['kid_per']=count_df['count_kidney']/count_df['count']
count_df['CCRCC_per']=count_df['count_CCRCC']/count_df['count']
count_df['cyst_per']=count_df['count_cyst']/count_df['count']
```

```python
sns.distplot(count_df['kid_per'],label='kidney')
sns.distplot(count_df['CCRCC_per'],label='CCRCC')
plt.legend()
```

```python
sns.jointplot(x='cyst_per',y='CCRCC_per',data=count_df)
```

```python
per_cols=[col for col in count_df.columns if 'per' in col]
g=sns.pairplot(count_df[per_cols])
plt.subplots_adjust(top=0.9)
plt.suptitle('Lesion Pair Plot',size=20)
plt.rcParams["font.size"] = 18
plt.rcParams["font.weight"] = 600

plt.tight_layout()
plt.savefig('./lesionLpairplot.png')
plt.show()
```
